```python
yd['published_month'] = yd['published_at'].dt.month 
yd.head(2)
```

```python
# Just to seperate Date and Time in 2 columns 
yd['published_at'] = yd.published_at.astype('string')
yd.published_at
yd['date'] = yd['published_at'].str.split().str[0]
yd['time'] = yd['published_at'].str.split().str[1]
yd.head(2)

yd['date'] = pd.to_datetime(yd.date)
yd['new_month'] = yd['date'].dt.month
yd.head(2)
```

```python
yd.loc[:,['view_count', 'dislikes']].corr()
```

```python
import matplotlib.pyplot as plt
plt.scatter(x = yd.view_count, y = yd.dislikes)
plt.xlabel('Views')
plt.ylabel('Dislikes')
```

```python
# Based on the above scatter plot and correlation value, it can be observed that there is a ~ 0.68 positive correlation 
# between the number of views and dislikes. This means that as the number of views increase, the number of 
# dislikes also tend to increase. 
```
